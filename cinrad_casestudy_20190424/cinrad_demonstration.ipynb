```python
def read_csv(csv_ffn, header_line):
    """
    CSV reader used for the radar locations file (comma delimited)
    
    Parameters:
    ===========
        csv_ffn: str
            Full filename to csv file
            
        header_line: int or None
            to use first line of csv as header = 0, use None to use column index
            
    Returns:
    ========
        as_dict: dict
            csv columns are dictionary
    
    """
    df = pandas.read_csv(csv_ffn, header=header_line)
    as_dict = df.to_dict(orient='list')
    return as_dict
```

```python
radar_volume = '/home/meso/dev/jss_notebooks/cinrad_casestudy_20190424/data/Z_RADR_I_Z9352_20190424094100_O_DOR_CB_CAP.bin.bz2' #these are *.bin files
data = pycwr.io.auto_io.radar_io(radar_volume)
PRD = data.ToPRD(withlatlon=True)
print(PRD.scan_info)
graph = pycwr.draw.SingleRadarPlotMap.RadarGraphMap(PRD)
graph.plot(1, "dBZ")
plt.show()
```
